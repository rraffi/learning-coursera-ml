```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f'x_train = {x_train}')
print(f'y_train = {y_train}')
```

```python
print('x_train.shape: f{x_train.shape}')
m = x_train.shape[0] # .shape return length of each dimension. In this case, x_train has only one-dimension.
print(f'Number of training examples is: {m}')
```

```python
#ith training data

i = 1
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plot training data
def show_chart():
    plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')
    plt.title("Housing Prices (Atlanta 2015)")
    plt.xlabel("Size (1000 sqft)")
    plt.ylabel("Price (per $1000)")
    plt.legend()
    plt.show()

show_chart()
```

```python
# Def of prediction model function for linear regression f = wx + b
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    
    for i in range(m):
        f_wb[i] = w * x[i] + b 
    
    return f_wb
```

```python
# Model parameter values for w and b

w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")

# Compute prediction model output

tmp_f_wb = compute_model_output(x_train, w, b)

# Plot prediction model
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')
show_chart()
```

```python
# Predictions

x_i = 1.2
cost_1200sqft = w * x_i + b

print(f"Cost of 1200sqft house will be: {cost_1200sqft:.0f} thousand dollars")
```
